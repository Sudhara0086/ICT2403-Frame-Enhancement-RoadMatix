```python
import cv2

video_path = "C:/Users/Sudhara/OneDrive/Desktop/Lab_Practicals/Img_Enhance_Meth_Project/Video/road_video.mp4"
cap = cv2.VideoCapture(video_path)

frame_count = 0
saved_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count % 5 == 0:
        cv2.imwrite("frame_" + str(saved_count) + ".jpg", frame)
        saved_count += 1

    frame_count += 1

cap.release()

print("Frames extracted:", saved_count)
```
