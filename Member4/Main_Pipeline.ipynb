```python
import cv2
import os
import numpy as np

input_folder = "Frames/"
final_output = "Final_Enhanced/"
os.makedirs(final_output, exist_ok=True)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))

def variance_of_laplacian(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()

threshold = 400

frame_files = sorted(
    [f for f in os.listdir(input_folder)
     if f.startswith("frame_") and f.endswith(".jpg")],
    key=lambda x: int(x.split("_")[1].split(".")[0])
)

for file in frame_files:

    img = cv2.imread(os.path.join(input_folder, file))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Noise Removal
    denoised = cv2.GaussianBlur(gray, (5,5), 0)

    # CLAHE
    enhanced = clahe.apply(denoised)

    # Blur Detection
    sharpness = variance_of_laplacian(enhanced)

    if sharpness < threshold:
        blurred = cv2.GaussianBlur(enhanced, (5,5), 0)
        enhanced = cv2.addWeighted(enhanced, 1.5, blurred, -0.5, 0)

    cv2.imwrite(os.path.join(final_output, file), enhanced)

print("Full pipeline enhancement completed.")
```
