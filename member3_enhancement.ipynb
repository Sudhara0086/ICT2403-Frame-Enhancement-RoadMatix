```python
# 1. GrayScale Conversion
import cv2
# Input folder
input_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/Frames (3)/"
# Output folder 
output_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/GrayScale_Frames/"

frame_count = 0
converted_count = 0

while True:
    # Read frame
    frame = cv2.imread(input_folder + "frame_" + str(frame_count) + ".jpg")
    
    if frame is None:
        break
    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Save grayscale image
    cv2.imwrite(output_folder + "frame_" + str(frame_count) + ".jpg", gray)
    converted_count += 1
    frame_count += 1

print("Grayscale Conversion done for", converted_count, "frames.")
```

```python
# 3. calculate sharpness for all frames..
import cv2

input_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/GrayScale_Frames/"

def variance_of_laplacian(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()

frame_count = 0
sharpness_values = []

while True:
    img = cv2.imread(input_folder + "frame_" + str(frame_count) + ".jpg", cv2.IMREAD_GRAYSCALE)
    if img is None:
        break

    sharpness = variance_of_laplacian(img)
    sharpness_values.append(sharpness)

    print(f"frame_{frame_count}.jpg - Sharpness: {sharpness:.2f}")

    frame_count += 1

print("\nSharpness calculated for", frame_count, "frames.")
print("Minimum Sharpness:", min(sharpness_values))
print("Maximum Sharpness:", max(sharpness_values))
```

```python
# 4. Detect Blurry Frames Using Threshold
import cv2

input_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/GrayScale_Frames/"

def variance_of_laplacian(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()

frame_count = 0
threshold = 400  
blurry_frames = []

while True:
    img = cv2.imread(input_folder + "frame_" + str(frame_count) + ".jpg", cv2.IMREAD_GRAYSCALE)
    if img is None:
        break

    sharpness = variance_of_laplacian(img)

    if sharpness < threshold:
        print(f"frame_{frame_count} - BLURRY ({sharpness:.2f})")
        blurry_frames.append(frame_count)

    frame_count += 1

print("\nTotal Blurry Frames:", len(blurry_frames))
print("Blurry Frame Numbers:", blurry_frames)
```

```python
# 5. Sharpen Blurry Frames
import cv2
import numpy as np

input_folder ="C:/Users/Acer/Desktop/Image_Processing_Project_New/GrayScale_Frames/"
output_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/Sharpened_Frames/"

# List of blurry frames from previous step
blurry_frames = [28, 29, 123, 124, 198]

for frame_no in blurry_frames:
    # Read the grayscale frame
    img = cv2.imread(input_folder + f"frame_{frame_no}.jpg", cv2.IMREAD_GRAYSCALE)
    if img is None:
        print("Frame not found:", frame_no)
        continue
    
    # Apply Gaussian blur
    blurred = cv2.GaussianBlur(img, (5,5), 0)
    
    # Unsharp mask: sharpened = original + (original - blurred)
    sharpened = cv2.addWeighted(img, 1.5, blurred, -0.5, 0)

    # Save the sharpened frame
    cv2.imwrite(output_folder + f"frame_{frame_no}_sharp.jpg", sharpened)

print("Sharpening done for all blurry frames.")
```

```python
# 6. Create Before-After Comparison Images
import cv2
import numpy as np

# Folders
grayscale_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/GrayScale_Frames/"
sharpened_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/Sharpened_Frames/"
comparison_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/Comparisons/"

# Blurry frames to compare
blurry_frames = [28, 29, 123, 124, 198]

for f in blurry_frames:
    # Read original and sharpened frames
    original = cv2.imread(grayscale_folder + f"frame_{f}.jpg", 0)
    sharp = cv2.imread(sharpened_folder + f"frame_{f}_sharp.jpg", 0)
    
    if original is None or sharp is None:
        continue  
    
    # Add simple labels using cv2.putText
    original_bgr = cv2.cvtColor(original, cv2.COLOR_GRAY2BGR)
    sharp_bgr = cv2.cvtColor(sharp, cv2.COLOR_GRAY2BGR)
    
    cv2.putText(original_bgr, "Original", (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
    cv2.putText(sharp_bgr, "Sharpened", (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
    
    # Combine images side by side
    combined = np.hstack((original_bgr, sharp_bgr))
    
    # Save the combined image
    cv2.imwrite(comparison_folder + f"comparison_{f}.jpg", combined)

print("All comparison images saved")
```

```python
# 7. Sharpness Before and After Sharpening
import cv2
import numpy as np

# Folder path
grayscale_folder = "C:/Users/Acer/Desktop/Image_Processing_Project_New/GrayScale_Frames/"

# Blurry frames detected earlier
blurry_frames = [28, 29, 123, 124, 198]

# Function to calculate sharpness
def variance_of_laplacian(image):
    return cv2.Laplacian(image, cv2.CV_64F).var()

print(f"{'Frame Name':<15} {'Before':<15} {'After':<15} {'Improvement':<15}")

for frame_no in blurry_frames:
    
    # Read grayscale image
    original = cv2.imread(grayscale_folder + "frame_" + str(frame_no) + ".jpg",cv2.IMREAD_GRAYSCALE)
    
    if original is None:
        print("Frame not found:", frame_no)
        continue

    # Sharpness before
    sharp_before = variance_of_laplacian(original)

    # Apply Unsharp Mask
    blurred = cv2.GaussianBlur(original, (5,5), 0)
    sharpened = cv2.addWeighted(original, 1.5, blurred, -0.5, 0)

    # Sharpness after
    sharp_after = variance_of_laplacian(sharpened)
    #calculate improvement (differences)
    improvement = sharp_after - sharp_before

    # Print 
    print(f"frame_{frame_no}.jpg       {sharp_before:<15.2f}     {sharp_after:<15.2f}     {improvement:<15.2f}")
```
