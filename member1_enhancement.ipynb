```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

frames_path = "C:/Users/HomePC/OneDrive/Desktop/enhancement/Frames/Frames"
output_folder = "Member1_Output/"
os.makedirs(output_folder, exist_ok=True)

frame_files = sorted(
    [f for f in os.listdir(frames_path)
     if f.startswith("frame_") and f.endswith(".jpg")],
    key=lambda x: int(x.split("_")[1].split(".")[0])
)

print("Total Frames Found:", len(frame_files))
```

```python
img = cv2.imread(os.path.join(frames_path, frame_files[0]))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Save grayscale example
plt.figure()
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.savefig(os.path.join(output_folder, "grayscale_example.jpg"))
plt.close()

# Save histogram before filtering
plt.figure()
plt.hist(gray.ravel(), bins=256)
plt.title("Histogram Before Filtering")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.savefig(os.path.join(output_folder, "histogram_before.jpg"))
plt.close()

print("Mean Intensity:", np.mean(gray))
print("Standard Deviation:", np.std(gray))
```

```python
gaussian = cv2.GaussianBlur(gray, (5,5), 0)
median = cv2.medianBlur(gray, 5)

# Save comparison image
comparison = np.hstack((gray, gaussian, median))
cv2.imwrite(os.path.join(output_folder, "comparison_example.jpg"), comparison)

# Save histograms after filtering
plt.figure()
plt.hist(gaussian.ravel(), bins=256)
plt.title("Histogram After Gaussian")
plt.savefig(os.path.join(output_folder, "histogram_gaussian.jpg"))
plt.close()

plt.figure()
plt.hist(median.ravel(), bins=256)
plt.title("Histogram After Median")
plt.savefig(os.path.join(output_folder, "histogram_median.jpg"))
plt.close()
```

```python
results = []

for frame_name in frame_files[:30]:

    img = cv2.imread(os.path.join(frames_path, frame_name))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    gaussian = cv2.GaussianBlur(gray, (5,5), 0)
    median = cv2.medianBlur(gray, 5)

    # Save filtered images
    cv2.imwrite(os.path.join(output_folder, "gaussian_" + frame_name), gaussian)
    cv2.imwrite(os.path.join(output_folder, "median_" + frame_name), median)

    # Calculate standard deviation
    original_std = np.std(gray)
    gaussian_std = np.std(gaussian)
    median_std = np.std(median)

    results.append([frame_name, original_std, gaussian_std, median_std])

print("Noise removal applied to 30 frames successfully.")
```

```python
print("\nNoise Reduction Results (Standard Deviation)")
print("Frame\tOriginal\tGaussian\tMedian")

for r in results:
    print(f"{r[0]}\t{r[1]:.4f}\t\t{r[2]:.4f}\t\t{r[3]:.4f}")

avg_original = np.mean([r[1] for r in results])
avg_gaussian = np.mean([r[2] for r in results])
avg_median = np.mean([r[3] for r in results])

print("\nAverage Standard Deviation:")
print("Original:", round(avg_original, 4))
print("Gaussian:", round(avg_gaussian, 4))
print("Median:", round(avg_median, 4))

print("\nFiltered images and analysis saved in Member1_Output folder.")
```
